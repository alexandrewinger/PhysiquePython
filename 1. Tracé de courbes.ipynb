# 1. Comment tracer une courbe avec Python?
---

Nous aurons besoin des bibliothèques:
1. `numpy`pour réaliser les [calculs numériques](https://numpy.org/doc/1.18/reference/index.html)
2. `matplotlib` et son [module](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.html) `pyplot`pour tracer des courbes.

```python
import numpy as np                 #on renomme les modules et leurs fonctions avec des raccourcis pour les appeler plus facilement dan le code.
import matplotlib.pyplot as plt
```

Imaginons que l'on veuille représenter la courbe représentative de la fonction:
$$ \normalsize{ y: t \mapsto y(t)=4sin(3t)+2}$$
définie sur l'intervalle $\normalsize{I= [0,2\pi[}$

Il faut d'abord créer une liste de valeurs de la variable t dans l'intervalle $I$ avec la commande `np.arange()`, que l'on affichera avec la fonction `print(objet)`.
Pour obtenir la valeur de pi =3.1415..., on importe la valeur depuis numpy avec la commande `numpy.pi`, qui s'écrit ici `np.pi`puisqu'on l'a renommé ainsi dans le préambule la bibliothèque `numpy`.

```python
t = np.arange(0, 2*np.pi)
print(t)
```

Par défaut, le **pas** (l'intervalle entre deux valeurs consécutives de la liste) est entier, mais on peut changer ça dans les options, en mettant un pas de 0.5 par exemple: 

```python
t = np.arange(0, 2*np.pi, 0.5)
print(t)
```

Essayons de calculer les valeurs correspondantes de y:

```python
y = 4*np.sin(3*t)+2
print(y)
```

Il suffit de représenter ces points sur un graphique en appelant la fonction `.plot`du module `matplotlib.pyplot`, que l'on a raccourci en `plt`:

```python
plt.plot(t, y)
plt.show() # obligatoire pour afficher le graphique via python, mais superflu ici. 
```

Ca n'est pas tout à fait la représentation d'une fonction sinusoïdale, car notre pas est trop grand. Changeons ça en créant une variable `pas`:

```python
pas = 0.01
t = np.arange(0, 2*np.pi, pas)
y = 4*np.sin(3*t)+2
plt.plot(t, y)
plt.show()
```

Et voilà!

On peut ajouter un titre, des légendes aux axes, changer le domaine d'affichage ...

```python
plt.figure(figsize=(10,6))             # à placer avant le plot, pour changer la taille d'affichage (unités en pouces par défaut)
plt.plot(t, y, linestyle='-.', color = 'darkred', linewidth = 3, label='y(t)') #linewidth = largeur de ligne
plt.xlabel('Temps t (en s)')           # ajoute un tire à l'axe des abscisses
plt.ylabel('Fonction y')               # ajoute un titre à l'axe des ordonnées
plt.title('y en fonction du temps')    # ajoute un titre au graphique
plt.legend()                           # affiche les légendes spécifiées dans les arguments du plot (ici label='y(t)')
plt.xlim([-1, 8])                      # affiche x entre -1 et 8
plt.ylim([-6,6])                       # affiche y entre -6 et 6
plt.grid(True)                         # affiche une grille
plt.show()
```

Vous trouverez ici [la liste des couleurs](https://matplotlib.org/3.1.0/gallery/color/named_colors.html) et là [la liste des styles de lignes](https://matplotlib.org/3.1.0/gallery/lines_bars_and_markers/linestyles.html) disponibles.

# Le code en résumé :

```python

début=0  #borne inférieure de l'ensemble de définition de la fonction à tracer (incluse)
fin=10   #borne supérieure de l'ensemble de définition de la fonction à tracer (exclue)
pas = 0.01
t = np.arange(début, fin, pas)
y = 3*t**2-5*t+30 #expression de la fonction à étudier

plt.figure(figsize=(15,6)) 
plt.plot(t, y, linestyle='dotted', color = 'blueviolet', linewidth = 3, label='y(t)')
plt.xlabel('Temps t (en s)')
plt.ylabel('Fonction y')
plt.title('y en fonction du temps')
plt.legend() 
plt.xlim([début, fin])
plt.ylim([0,300])
plt.grid(True) 
plt.show()
```
